```python
col_list = []

for i in churn_data.columns:
  if((churn_data[i].dtype == "object")&(i!="Churn")):
    col_list.append(i)


print(col_list)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for i in col_list:
  churn_data[i] = le.fit_transform(churn_data[i])
```

```python


x = churn_data.iloc[:,:-1]        # features or the independent columns
y = churn_data["Churn"]     # target or the dependent column
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=9999)
```

```python
x_train   
```

```python
x_test
```

```python


classifier1 = RandomForestClassifier(n_estimators=100)   # no of decision trees = 100

classifier1.fit(x_train,y_train) #training the model
```

```python
y_pred = classifier1.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_pred,y_test)
```
